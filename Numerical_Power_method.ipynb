##**Power Method and Inverse Power Method**

Import required libraries

```python
import numpy as np
```

Defining the normalization function which can be used for both power method and inverse power method.

```python
def normalize(x):
    fac = abs(x).max()
    if fac!=max(x):
      fac = -fac
    x_n = x / fac
    return fac, x_n
```

Define the power method function which can be used for power method.

```python


def power_method(a,x0,n):
    print(f"The actual eigen values are : \n{np.linalg.eig(a)[0]}")
    print(f"\nThe actual eigen vectors are : \n{np.linalg.eig(a)[1]}")
    print("\nEigen values and vectors at each iteration using power method")

    for i in range(n):
        
        x0 = np.dot(a, x0)
        lambda_1, x0 = normalize(x0)
        
        print(f"\n***********Iteration - {i+1}***********")
        print('\nEigenvalue:', lambda_1)
        print('\nEigenvector:', x0)
        



```

**1 (d)** . **Find the first three iterations obtained by the Power method applied to the following matrix**

$$
\left[\begin{array}{rrrr}
4 & 1 & 1 & 1 \\
1 & 3 & -1 & 1 \\
1 & -1 & 2 & 0 \\
1 & 1 & 0 & 2
\end{array}\right]
$$
$\operatorname{Use}{X}^{(0)}=(1,-2,0,3)^{t}$.

```python
x0 = np.array([1,-2,0,3 ])
a = [[4,1,1,1], 
    [1,3,-1,1],
    [1,-1,2,0],
    [1,1,0,2]]
n= 3
power_method(a,x0,n)
```

```python
def inverse_power_method(a,x0,n):
    q = (np.dot(np.transpose(x0),(np.dot(a,x0))))/(np.dot(np.transpose(x0),x0))
    print("\n q value is : ",q)
    a = np.subtract(a,q*np.eye(4,4))
    print(f"\n The actual eigen values are : {np.linalg.eig(a)[0]}")
    print("\nEigen values and vectors at each iteration using Inverse power method :\n")
    a_inv = np.linalg.inv(a)
    a_inv = np.array(a_inv)
    print(a_inv)
    x0 = np.array([1,-2,0,3 ])
    for i in range(3):
        x0 = np.dot(a_inv, x0)
        lambda_1, x0 = normalize(x0)
        mu = (1/lambda_1) + q
        print("\n")
        print(f"***********Iteration - {i+1}***********")
        print('\nEigenvalue:', mu)
        print('\nEigenvector:', x0)
    
```
